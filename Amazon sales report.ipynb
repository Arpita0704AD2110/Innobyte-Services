**Importing libraries**

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
````

**Importing dataset**

````python
#importing the dataset of amazon sales report
data = pd.read_csv("Amazon Sale Report.csv")
````

````python
#converting teh dataset to the dataframes
data = pd.DataFrame(data)
data
````

**Exporatory data analysis (EDA)**

````python
#Deleting the index column from the dataset
data = data.drop(columns = ['index'])
data
````

````python
#Finding the sum of all null values in different coulmns
data.isnull().sum()
````

````python
#Deleting the column new, pendings, fulfilllment from the dataset
data = data.drop(columns = ['New','PendingS', 'fulfilled-by'])
data
````

````python
#finding the rows and columnn of the duplicated data
data[data.duplicated()].shape
````

````python
#new deleting all the duplicated rows
data = data.drop_duplicates()
data
````

````python
#Accessing only those orders which which contain atleast 1 quantity of order
data = data[(data['Qty'] >= 1)]
data
````

````python
#Now checking the sum of all null values in different columns
data.isnull().sum()
````

````python
#deleting the rows with the combination of two columns i.e. amount and currency
#because they contain same number of null values
data = data.dropna(subset = ['Amount', 'currency'])
#deleting the rows with the combination of ship-city, ship-state, ship-postalcode, ship-country columns
#because they contain same number of null values i.e 30
data = data.dropna(subset = ['ship-city','ship-state','ship-postal-code','ship-country'])
data
````

````python
#now we can see that it doesn't contain the null values
data.isnull().sum()
````

> **1. Sales Overview: Understand the overall sales performance, trends, and patterns over time**

````python
#convert the Date column to the type of date time format
data['Date'] = pd.to_datetime(data['Date'], format = "%m-%d-%y", errors='coerce')
````

````python
#group the sum of Amount and count of orderid by the date of order
sales = data.groupby('Date').agg({'Amount' : 'sum', 'Order ID' : 'count'}).reset_index()
````

````python
#plot the graph for above
plt.figure(figsize =[18,5])
plt.subplot(1,2,1)
plt.plot(sales['Date'], sales['Amount'], color = 'red')
plt.xlabel("Date",labelpad= 15)
plt.ylabel('Amount', labelpad= 10)
plt.title("Sum of Amount by order date",fontdict =  {'fontsize': 15,'color': '#030202'}, pad =15)


plt.subplot(1,2,2)
plt.plot(sales['Date'], sales['Order ID'],color='blue')
plt.xlabel('Date', labelpad= 15)
plt.ylabel('No of order', labelpad= 10)
plt.title("Count of order Id by order date",fontdict =  {'fontsize': 15,'color': '#030202'}, pad = 15)
plt.show()
````

````python
#plotting the graph having opposite y axis as amount and order id and a single x axis as date
fig,axis1 = plt.subplots(figsize = [10,5]) 
axis1.plot(sales['Date'], sales['Amount'], color = 'blue', marker = 'o', ls = '--') 
axis2=axis1.twinx() 
axis2.plot(sales['Date'], sales['Order ID'], color = 'red', marker = 'o', ls = '--') 
axis1.set_title('Sales over time', pad = 20, fontdict =  {'fontsize': 20,'color': '#030202'})
axis1.set_xlabel('Date of order', labelpad=15,fontdict =  {'fontsize': 12, 'color' : '#292727'})
axis1.set_ylabel('Amount of order', labelpad=15,fontdict =  {'fontsize': 12 , 'color' : '#292727'})
axis2.set_ylabel('Number of order', labelpad=15,fontdict =  {'fontsize': 12 , 'color' : '#292727'})
plt.show()
````

````python
#total sales
total_sales = data['Amount'].sum()
# Average sales per order
avg_sales_per_order = total_sales / (data['Order ID'].nunique())
print(f"Total Sales: {total_sales}")
print(f"Average Sales per Order: {avg_sales_per_order:.2f}")
````

> **2. Product Analysis: Analyze the distribution of product categories, sizes, and quantities sold to identify popular
products**

````python
#plotting multiple graph in once
plt.figure(figsize = [20,6])
plt.subplot(1,3,1)
#Count the occurance of differnt category
data1 = data['Category'].value_counts()
plt.title("Distribution of product categories")
plt.xticks(rotation=45, ha='right')
#plotting graph using seaborn library
a = sns.barplot(data=data1, gap = 0.3, color = 'Blue')
#using for loop to print the data lael value on the bar
for i in a.containers:
    a.bar_label(i)

plt.subplot(1,3,2)
#count the occurance of different sizes
data2 = data['Size'].value_counts()
plt.title("Distribution of product sizes")
b = sns.barplot(data = data2, gap = 0.3, color = "red")
for i in b.containers:
    b.bar_label(i)

plt.subplot(1,3,3)
#count the occurance of fiffernt quantity
data3 = data['Qty'].value_counts()
plt.title("Distribution of product quantity")
c = sns.barplot(data = data3, gap = 0.3, color = "green")
for i in c.containers:
    c.bar_label(i)
plt.show()
````

````python
# Group by product category and calculate total sales and order count
category_sales = data.groupby('Category').agg({'Amount': 'sum', 'Order ID': 'count'}).reset_index()
# Group by product size and calculate total sales and order count
size_sales = data.groupby('Size').agg({'Amount': 'sum', 'Order ID': 'count'}).reset_index()
plt.figure(figsize = [20,5])
# Plot the distribution of product categories
plt.subplot(1,2,1)
sns.barplot(data=category_sales, x='Category', y='Amount', palette='viridis')
plt.title('Total Sales by Product Category', pad = 15,fontdict =  {'fontsize': 18,'color': '#030202'})
plt.xlabel('Product Category')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
# Plot the distribution of product sizes
plt.subplot(1,2,2)
sns.barplot(data=size_sales, x='Size', y='Amount', palette='viridis')
plt.title('Total Sales by Product Size', pad = 15,fontdict =  {'fontsize': 18,'color': '#030202'})
plt.xlabel('Product Size')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.show()
````

> **3. Fulfillment Analysis: Investigate the fulfillment methods used and their effectiveness in delivering orders**

````python
#count the occurance of fulfilment
data4 = data['Fulfilment'].value_counts()
print(data4, "\n"*2)
print("Percentage of Amazon :- ",(data4[0]/len(data))*100)
print("Percentage of merchant :- ",(data4[1]/len(data))*100)
````

````python
#pie chart to represent the fulfilment methods
plt.pie(data4.values, labels = data4.index, autopct = '%.2f%%', shadow = True, startangle = 30, explode= [0,0.1])
plt.title("Distribution of fulfilment method")
plt.legend(bbox_to_anchor = [0.7,0.5,0.5,0.5])
plt.show()
````

````python
#count the occurance of category in different fulfilment method
data5 = data.groupby('Category')['Fulfilment'].value_counts()
data5 = pd.DataFrame(data)
plt.figure(figsize = [15,5])
a = sns.countplot(data = data5, x = data5['Category'], hue = data5['Fulfilment'], palette = "Dark2")
for i in a.containers:
    a.bar_label(i)
plt.title("Distribution of fulfilment method for each product Category")
plt.show()
````

````python
#calculate the size of order status with respectto fulfilment method
#And fill the null values with 0 in the data
data6 = data.groupby(['Fulfilment','Status']).size().unstack().fillna(0)
data6
````

````python
plt.figure(figsize = [5,15])
data6.plot(kind = 'bar', stacked = True, figsize = [10,5])
plt.title("Distribution of fulfillment method for every order status")
plt.legend(bbox_to_anchor = [0.5,0.4,0.5,0.5], ncols = 2)
plt.xlabel('Fulfillment method')
plt.ylabel('Numbers of orders')
plt.show()
````

> **4. Customer Segmentation: Segment customers based on their buying behaviour, location, and other relevant
factors**

````python
# Group by 'Order ID' to calculate total amount and frequency for each customer
data7 = data.groupby('Order ID').agg({ 'Amount': 'sum',    'Order ID': 'count' }).rename(columns={'Order ID': 'Order Frequency'}).reset_index()
# Plot Total Spending vs. Order Frequency (Buying Behavior)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data7, x='Order Frequency', y='Amount', hue='Amount', palette='viridis')
plt.title('Customer Segmentation by Buying Behavior' , pad = 20, fontdict =  {'fontsize': 18,'color': '#030202'})
plt.xlabel('Order Frequency', labelpad=15,fontdict =  {'fontsize': 12, 'color' : '#292727'})
plt.ylabel('Amount spending', labelpad=15,fontdict =  {'fontsize': 12, 'color' : '#292727'})
plt.show()
````

````python
#count the occurance of ship-state
data8 = data['ship-state'].value_counts()
states = data8.index
no_of_orders = data8.values
#then convert them into dataframe
data9= pd.DataFrame([states,no_of_orders])
data9
````

````python
plt.figure(figsize = [10,25])
a = sns.barplot(data = data8, orient ='h', color = "Hotpink")
plt.title("Distribution of state count",fontdict =  {'fontsize': 18,'color': '#030202'}, pad = 20)
for i in a.containers:
    a.bar_label(i)
plt.show()
````

> **5. Geographical Analysis: Explore the geographical distribution of sales, focusing on states and cities.**

````python
#calculate the sum of amount with respect to ship-city and ship-state
sales1 = data.groupby(['ship-state', 'ship-city']).agg({'Amount':'sum'}).reset_index()
sales1
````

````python
data10 = sales1.groupby('ship-state').agg({'Amount':'sum'}).reset_index()
data10 = data10.sort_values(by ='Amount', ascending = False)

data11 = sales1.groupby('ship-city').agg({'Amount':'sum'}).reset_index()
data11 = data11.sort_values(by = 'Amount', ascending = False)
````

````python
plt.figure(figsize = [18,5])
sns.barplot(data = data10, x = 'ship-state', y= 'Amount', log = True)
plt.xticks(rotation = 90)
plt.title("Distribution of total amount in various states",fontdict =  {'fontsize': 20,'color': '#030202'}, pad = 20)
plt.show()
````

````python
plt.figure(figsize = [18,5])
sns.barplot(data = data11.head(67), x = 'ship-city', y = 'Amount', log = True)
plt.title("Distribution of total amount in various city",fontdict =  {'fontsize': 20,'color': '#030202'}, pad = 20)
plt.xticks(rotation = 90)
plt.show()
````

````python
a = data10.head(10)
b = data11.head(10)
plt.figure(figsize = [15,5])
plt.subplot(1,2,1)
sns.barplot(data = a, x = 'ship-state', y= 'Amount', log = True)
plt.title("Top 10 state by total sales",fontdict =  {'fontsize': 15,'color': '#030202'}, pad = 15)
plt.xticks(rotation = 45)
plt.subplot(1,2,2)
sns.barplot(data = b, x = 'ship-city', y = 'Amount', log = True)
plt.title("Top 10 city by total sales",fontdict =  {'fontsize': 15,'color': '#030202'}, pad = 15)
plt.xticks(rotation = 45)
plt.show()
````

````python
#assuming the amount to be on same axis and state and cityon opposite y axis
data10 = data10.head(50)
data11 = data11.head(50)
fig,axis1 = plt.subplots(figsize = [10,15]) 
axis1.plot(data10['Amount'], data10['ship-state'], color = 'blue') 
axis2=axis1.twinx() 
axis2.plot(data11['Amount'], data11['ship-city'], color = 'red') 
axis1.set_title("Top 50 states and cities with total amount",fontdict =  {'fontsize': 15,'color': '#030202'}, pad = 15)
axis1.set_ylabel("Ship-state", labelpad = 20)
axis2.set_ylabel("Ship-city", labelpad = 20)
axis1.set_xlabel("Amount", labelpad = 20)

plt.show()
````

> **6. Business Insights: Provide actionable insights and recommendations based on the analysis to optimize sales
strategies, improve customer satisfaction, and enhance overall business performance**

````python
new_data = data['ship-state'].unique().tolist()[1]
print("The top 1 state of the sales is ",new_data)
new_data1 = pd.DataFrame(data[data['ship-state'] == new_data].groupby('Size')['Category'].value_counts())
````

````python
new_data1
````

````python
plt.figure(figsize = [15,15])
x1 =[]
y = []
states = data['ship-state'].unique().tolist()
for new_data in states:
    new_data1 = pd.DataFrame(data[data['ship-state'] == new_data].groupby('Size')['Category'].value_counts())
    x = pd.DataFrame(new_data1.iloc[np.argmax(new_data1)])
    y.append(new_data1.iloc[np.argmax(new_data1)][0])
    x1.append(f'{new_data}:{x.columns.values[0][0]}, {x.columns.values[0][1]}')

a = sns.barplot(y = x1, x=y)
for i in a.containers:
    a.bar_label(i)
plt.show()
````
